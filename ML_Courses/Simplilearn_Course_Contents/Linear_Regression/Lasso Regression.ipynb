```python
# First, define X and y by loading your dataset
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset
df = pd.read_csv('Hitters.csv')

# Check the available columns in the dataset
# print(df.columns)

# Define X (features) and y (target variable)
# Assuming 'Salary' is the target column in the Hitters dataset
X = df.drop('Salary', axis=1)  # Replace with actual target column from your dataset
y = df['Salary']  # Replace with actual target column from your dataset

# Now split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Perform Standardization on numerical features.

from sklearn.preprocessing import StandardScaler

# Define list_numerical with your actual numerical column names from the DataFrame
# First, check what columns are available in your DataFrame
print("Available columns:", X_train.columns.tolist())

# Then use the actual column names that exist in your DataFrame
# For example, if your numerical columns are 'age', 'income', and 'score':
list_numerical = ['age', 'income', 'score']  # Replace with your actual numerical column names

# Make sure these columns exist before proceeding
for col in list_numerical:
    if col not in X_train.columns:
        print(f"Warning: Column '{col}' not found in DataFrame")

# Proceed with standardization if columns exist
scaler = StandardScaler().fit(X_train[list_numerical])
X_train[list_numerical] = scaler.transform(X_train[list_numerical])
X_test[list_numerical] = scaler.transform(X_test[list_numerical])
```

```python
X_train.head()
```

```python
# Imputing the values NA values in the series with the median
median_salary = y_train.median(skipna=True)
y_train = y_train.fillna(median_salary)
y_test = y_test.fillna(median_salary)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso

# Create dummy variables
dms = pd.get_dummies(
    df[['League', 'Division', 'NewLeague']],
    drop_first=True
)

# Target variable
y = df['Salary']

# Feature variables
X_ = df.drop(
    ['Unnamed: 0', 'Salary', 'League', 'Division', 'NewLeague'],
    axis=1
).astype('float64')

# Combine numerical + dummy variables
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)

# Drop rows with NaN values in either X or y
# This ensures both features and target are clean of missing values
mask = ~(X.isna().any(axis=1) | y.isna())
X = X[mask]
y = y[mask]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Lasso model
lasso_model = Lasso().fit(X_train, y_train)

# Intercept
lasso_model.intercept_
```

```python
print(lasso_model.coef_)
```

```python
# Training data
pred_train = lasso_model.predict(X_train)
mse_train = mean_squared_error(y_train, pred_train)
print('MSE training set', round(mse_train, 2))
print('R squared training set', round(r2_score(y_train, pred_train)*100,2))
print("---------------------------")

# Test data
pred_test = lasso_model.predict(X_test)
mse_test =mean_squared_error(y_test, pred_test)
print('MSE test set', round(mse_test, 2))
print('R squared test set', round(r2_score(y_test, pred_test)*100,2))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Lasso, LassoCV
import numpy as np
import pandas as pd

# Prediction using Lasso model
y_pred = lasso_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
r2_score(y_test, y_pred)
```

```python
# Lasso with Cross Validation
lasso_cv_model = LassoCV(
    alphas=np.random.randint(0, 1000, 100),
    cv=10,
    max_iter=10000,
    n_jobs=-1
).fit(X_train, y_train)
```

```python
# Tuned Lasso model
lasso_tuned = Lasso().set_params(alpha=14).fit(X_train, y_train)

y_pred_tuned = lasso_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred_tuned))
```
